```python
visas_df = pd.read_csv('no_of_visa_417_subclass_by_citizenship_country.csv')

visas_df
```

```python
pop_df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_4770387.csv')

pop_df
```

```python
# remove all columns except 2021 in pop df

pop_df = pop_df[['Country Name', '2021']]
```

```python
pop_df
```

```python
# join the two dataframes

merged_df = pd.merge(visas_df, pop_df, left_on='Citizenship Country', right_on='Country Name')

merged_df
```

```python
# drop '30/06/2021' column

merged_df.columns = ["Country", "Visas_2021", "Visas_2022", "Change", "Change_Rate", "Country_Name", "Population"]
```

```python
merged_df = merged_df[['Country','Visas_2022','Population']]
```

```python
merged_df
```

```python
merged_df["Visas_adjusted"] = merged_df["Visas_2022"] / merged_df["Population"] * 1000000
```

```python
merged_df
```

```python
merged_df = merged_df.sort_values(by=['Visas_adjusted'], ascending=False).head(10)
```

```python
merged_df
```

```python
%%R -i merged_df

# EXAMPLE R CELL

plt <- ggplot(merged_df) +
    geom_bar(aes(Visas_adjusted), stat="identity", y = "Country") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
    labs(x = "Visas granted per million people in country of citizenship", y = "Citizenship", title = "Top 10 countries with the most visas for 2022")

plt
```

```python
%%R 

irl_df <- merged_df %>%
    filter(Country == "Ireland") 

ggplot(merged_df) +
    geom_bar(aes(x=Visas_adjusted, y=reorder(Country, Visas_adjusted) fill=..Country..=="Ireland"), stat="identity") +
    theme_minimal() +
    labs(y = "Citizenship country of applicant", x = "417 visas granted per million", title = "title")
```

```python
%%R

ggplot(merged_df) +
    geom_bar(aes(x=Visas_adjusted, y=reorder(Country, Visas_adjusted)), stat="identity") +
    theme_minimal() +
    labs(y = "Citizenship country of applicant", x = "417 visas granted per million", title = "title")
```

```python
# make a new column in merged df that is equal to 0 or 1 depending on if the country is Ireland or not

merged_df["Ireland"] = np.where(merged_df["Country"] == "Ireland", 1, 0)
```

```python
merged_df
```

```python
%%R -i merged_df

ggplot(merged_df) +
    geom_bar(aes(x=Visas_adjusted, y=reorder(Country, Visas_adjusted), fill=Ireland), stat="identity") +
    theme_minimal() +
    labs(y = "Citizenship country of applicant", x = "417 visas granted per million", title = "title")
```
